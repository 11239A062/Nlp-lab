```python
from collections import Counter

text = input("Enter your text: ").lower().split()
trigram = [(text[i], text[i+1], text[i+2]) for i in range(len(text)-2)]
count = Counter(trigram)

print("\nTrigram count:")
for tri, c in count.items():
    print(tri, ":", c)
```
