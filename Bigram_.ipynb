```python
from collections import Counter

t = input("Enter text: ").lower().split()
b = [(t[i], t[i+1]) for i in range(len(t)-1)]

wc, bc = Counter(t), Counter(b)

print("\nBigram probabilities:")
for (w1, w2), c in bc.items():
    print(f"{w1}, {w2}: {c / wc[w1]:.4f}")
```
