```python
import spacy

# Load small English model
nlp = spacy.load("en_core_web_sm")

# Example sentence
text = "The cats are playing happily."

# Process text
doc = nlp(text)

# Print morphology info
for token in doc:
    print(token.text, "→ Lemma:", token.lemma_, "| Morph:", token.morph)
```
