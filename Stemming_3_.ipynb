```python
from nltk.stem import PorterStemmer

# Create stemmer object
stemmer = PorterStemmer()

# Sample words
words = ["running", "runs", "runner", "easily", "fairly"]

# Apply stemming
for word in words:
    print(word, "→", stemmer.stem(word))
```
