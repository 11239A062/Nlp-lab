```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

text = "ChatGPT generates useful code snippets."
doc = nlp(text)

# Print word and POS
for token in doc:
    print(token.text, token.pos_, token.tag_)
```
